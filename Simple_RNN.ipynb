# 🧠 Simple RNN (Recurrent Neural Network)

---

# 📌 Introduction to Simple RNN

A **Simple Recurrent Neural Network (Simple RNN)** is a type of neural network specially designed for **sequential data**.

Unlike traditional neural networks, RNNs can **remember previous information** using a memory called a **hidden state**.

This helps the model understand data where **order matters**.

---

# 🤔 Why Do We Need RNN?

Normal Neural Networks cannot remember previous inputs.

For example:

Sentence:
```text
"I love learning AI"
```

To understand the word **"AI"**, the model should remember:

- "I"
- "love"
- "learning"

Simple RNN solves this problem by storing previous information.

# 🏗️ Architecture of Simple RNN

A Simple RNN contains:

1. Input Layer
2. Hidden Layer (Memory)
3. Output Layer

---

# 🔄 Basic Flow of RNN

```text
Input (Xt) ----> Hidden State (Ht) ----> Output (Yt)
                      ↑
             Previous Hidden State (Ht-1)
```

---

# 🧠 Important Idea of RNN

The hidden state acts like **memory**.

At every time step:

- Current input is taken
- Previous memory is combined
- New memory is generated

---

# 📐 Mathematical Formula

```text
Ht = tanh(WxXt + WhHt-1 + b)
```

Where:

| Symbol | Meaning |
|---|---|
| Xt | Current input |
| Ht-1 | Previous hidden state |
| Ht | Current hidden state |
| Wx | Input weight |
| Wh | Hidden weight |
| b | Bias |

# 🧱 Structure of RNN Over Time

```text
X1 → [RNN] → H1
            ↓
X2 → [RNN] → H2
            ↓
X3 → [RNN] → H3
```

Where:

- X1, X2, X3 are inputs
- H1, H2, H3 are hidden states

# 🔥 Advantages of Simple RNN

✅ Good for sequential data  
✅ Remembers previous inputs  
✅ Useful in NLP tasks  
✅ Easy architecture for beginners  

---

# ❌ Limitations of Simple RNN

❌ Cannot remember very long sequences  
❌ Suffers from Vanishing Gradient Problem  
❌ Training becomes difficult for long text  

---

# 🚀 Advanced Versions of RNN

To overcome limitations:

| Model | Full Form |
|---|---|
| LSTM | Long Short-Term Memory |
| GRU | Gated Recurrent Unit |

These are improved versions of RNN.

# 🐍 Python Implementation of Simple RNN

## 📌 Import Libraries

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

---

# 📌 Create Model

```python
model = Sequential()

model.add(SimpleRNN(
    units=32,
    input_shape=(10,1)
))

model.add(Dense(1))
```

---

# 📌 Compile Model

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

---

# 📌 Display Model Summary

```python
model.summary()
```

---

# 📌 Explanation of Code

| Code | Meaning |
|---|---|
| SimpleRNN(32) | 32 neurons in RNN layer |
| input_shape=(10,1) | Sequence length = 10 |
| Dense(1) | Output layer |
| adam | Optimizer |
| mse | Mean Squared Error loss |

````python
# now load the sample data
df=pd.read_csv('/content/spam.csv',encoding='latin1')
````

````python
# removing the unamed columns like 2,3,4
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
````

````python
df=df.rename({'v1':'target','v2':'email'},axis=1) # renaming the columns like target and email
````

````python
# lets change the target column into numerical variables
df['target']=df['target'].map({'ham':0,'spam':1})
````

````python
# now lets divide the data into independent variable and in to dependent variable
x=df['target']
y=df['email']
````

````python
# importing nltk
import nltk
nltk.download('stopwords')
````

````python
nltk.download('wordnet')
````

````python
# now we can apply the nlp pipeline for each mail like lowering ,removing puncutations,stopworde
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
cleaned_text=[]
for i in df['email']:
  text=i.lower() #lowering data
  text=''.join([i for i in text if i not in string.punctuation]) # removed punctuations
  text = ' '.join([lem.lemmatize(k) for k in text.split() if k not in stopwords.words("english")])
  cleaned_text.append(text)
````
